# Question
Using Exclusive OR

Is \((p \oplus q) \oplus r \equiv p \oplus (q \oplus r)\)? Justify your answer.

They are logically equal. They both evaluate to true if either only one of \(p\), \(q\), or \(r\) is true, but not more than one. But, also they both evaluate to false if all three are true or all three are false.

\begin{enumerate}
    \item[(a)] \text{Distributive: } \((p \land \neg q) \lor p \equiv (p \lor p) \land (p \lor \neg q)\) \\
    Distributing \(p\) across the OR operation.

    \item[(b)] \text{Idempotent Law: } \((p \lor p) \equiv p\) \\
    Any proposition OR-ed with itself is logically equivalent to itself.

    \item[(c)] \text{Absorption Law: } \((p \lor \neg q) \land (p \lor q) \equiv p\) \\
    Whenever \(p\) is true, the whole expression evaluates to true.
\end{enumerate}

Yes, \((p \land \neg q) \lor p \equiv p\).
