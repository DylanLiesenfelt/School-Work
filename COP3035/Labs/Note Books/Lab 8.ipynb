# Exercise 2:

Implement a function to upload data from the .csv file storing in a list of dictionaries for each date using the
following format for each line: See example:


```python
file = r'../%Misc/SCHD.csv'

def readData(file):

    dataEntries = []

    with open(file) as file:
        headers = file.readline().strip().split(',')

        for line in file:
            lineData = line.strip().split(',')
            dataDict = {}

            for i in range(len(headers)):
                dataDict[headers[i]] = lineData[i]

            dataEntries.append(dataDict)

    return dataEntries

print(readData(file))
```

# Exercise 3: 
Implement a function that clean the data by converting string values to their appropriate numeric types.

```python
data = readData(file)

def cleanData(data):

    for entry in data:
        for point in entry:
            if isinstance(entry[point], str) and '2023-' not in entry[point] and '2024-' not in entry[point]:
                entry[point] = float(entry[point])

    return data

print(cleanData(data))
```

# Exercise 4:
Implement a function that calculates the daily change and percentage change for each stock entry.

```python
cleanedData = cleanData(data)

def calcChange(clean):
    dates = []
    opens = []
    closes = []

    for entry in clean:
        dates.append(entry['Date'])
        opens.append(entry['Open'])
        closes.append(entry['Close'])

    dailyChanges = []
    dailyPercent = []
    for i in range(len(closes)):
        change = closes[i] - opens[i]
        dailyChanges.append(change)

        percentChange = (change / opens[i]) * 100
        dailyPercent.append(percentChange)

    changes = []
    for j in range(len(dailyChanges)):
        changeDict = { 'Date' : dates[j],
                      'Change': dailyChanges[j],
                      'Percent': dailyPercent[j]}
        changes.append(changeDict)

    return changes

print(calcChange(cleanedData))
```

# Exercise 5:
Implement a function that calculates the average trading volume of the stock.

```python
def calcAvgVol(clean):
    total = 0
    count = 0

    for entry in clean:
        for point in entry:
            if point == 'Volume':
                total += entry[point]
            count += 1

    return total/count

calcAvgVol(cleanedData)
```
